

define functions


```python
def scatter_fit_web(var):
    plt.scatter(df[var],dep, label="data")
    plt.scatter(df[var],ypred, label="model", alpha=0.5,color='red')
    plt.ylabel('Web Transactions',fontsize=12)
    plt.xlabel(var,fontsize=12)
    plt.title('Web Transactions & '+var+', Deviations from the mean')
    plt.legend(loc='best')
    plt.show()
    
def scatter_fit_retail(var):
    plt.scatter(df[var],dep, label="data")
    plt.scatter(df[var],ypred, label="model", alpha=0.5,color='pink')
    plt.ylabel('Retail Transactions',fontsize=12)
    plt.xlabel(var,fontsize=12)
    plt.title('Retail Transactions & '+var+', Deviations from the mean')
    plt.legend(loc='best')
    plt.show()

def hist_plot(var):
    plt.hist(df[var],50,density=True,facecolor='g', alpha=0.75, histtype='bar', ec='black')
    plt.title('Histogram of '+var)
    plt.ylabel('Probability',fontsize=12)
    plt.xlabel(var,fontsize=12)
    current_values = plt.gca().get_xticks()
    plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in current_values])
    plt.grid(True, color = "grey", linewidth = "0.2", linestyle = "-.")
    plt.show()
    
def hist_plot_perc(var):
    n, bin_edges = np.histogram(df[var], 100)
    # Normalize it, so that every bins value gives the probability of that bin
    bin_probability = n/float(n.sum())
    # Get the mid points of every bin
    bin_middles = (bin_edges[1:]+bin_edges[:-1])/2.
    # Compute the bin-width
    bin_width = bin_edges[1]-bin_edges[0]
    # Plot the histogram as a bar plot
    plt.bar(bin_middles, bin_probability, width=bin_width, ec='black')
    plt.title('Histogram of '+var)
    plt.ylabel('Probability',fontsize=12)
    plt.xlabel(var,fontsize=12)
    current_values = plt.gca().get_xticks()
    plt.gca().set_xticklabels(['{:,.2f}'.format(x) for x in current_values])
    plt.grid(True, color = "grey", linewidth = "0.2", linestyle = "-.")
    plt.show()
    
def freq_plot(var):
    if df[var].dtype==float:
        country_count = df[var].value_counts()
        country_count.index=country_count.index.astype(int)
    elif df[var].dtype!="O" and df[var].dtype=="Int64":
        country_count = df[var].value_counts()
    else:
        country_count = df[var].fillna('Missing').value_counts(dropna=False)
    sns.set(style="darkgrid")
    sns.barplot(country_count.index[0:20],country_count.values[0:20],alpha=0.9)
    plt.title('Frequency Distribution of '+var)
    plt.ylabel('Number of Occurrences',fontsize=12)
    plt.xlabel(var,fontsize=12)
    plt.show()
    
def freq_plot_rot(var):
    if df[var].dtype==float:
        country_count = df[var].value_counts()
        country_count.index=country_count.index.astype(int)
    else:
        country_count = df[var].fillna('Missing').value_counts(dropna=False)
    sns.set(style="darkgrid")
    sns.barplot(country_count.index[0:20],country_count.values[0:20],alpha=0.9)
    plt.title('Frequency Distribution of '+var)
    plt.ylabel('Number of Occurrences',fontsize=12)
    plt.xlabel(var,fontsize=12)
    plt.xticks(rotation=45)
    plt.show()
```



load dataset


```python
df = pd.read_excel("Assessment Dataset - Feb 2022.xlsx")
print(df.head())
```

Data cleaning - data types

```python
df['Date']=pd.to_datetime(df['Date'],errors='coerce').dt.date
df['year'] = pd.DatetimeIndex(df['Date']).year
df['month'] = pd.DatetimeIndex(df['Date']).month
floats=['Number of Competitive Stores'
        ,'Percentage Discount',
        'Direct Mail Spend',	'Email Spend',	'TV Spend',	'Affiliate Spend',	'Online Ad Spend',
        'Facebook Ad Spend',	'Branded Search Spend',	'Unbranded Search Spend',	'Snapchat Spend',	
        'Online Video Spend'
        ]
ints=['Number of Owned Stores'
      ,'Retail Transactions',	'Web Transactions'
      ,'year'
      ,'month'
        ]
for i in floats:
    df[i]=pd.to_numeric(df[i],downcast="float",errors='coerce')
for i in ints:
    df[i]=df[i].astype("Int64",errors='ignore')
```



Visualise the data


```python
continuous=['Number of Competitive Stores','Direct Mail Spend','Email Spend','TV Spend',
    'Affiliate Spend','Online Ad Spend','Facebook Ad Spend','Branded Search Spend'
            ,'Unbranded Search Spend'
            ,'Snapchat Spend'
            ,'Online Video Spend'
            ,'Retail Transactions'
            ,'Web Transactions']
continuous_perc=['Percentage Discount']
discrete=['Number of Owned Stores','Promo'
          ]
discrete_rot=['Code'
          ]

for i in continuous:
    hist_plot(i)
for i in continuous_perc:
    hist_plot_perc(i)

for i in discrete:
    freq_plot(i)
for i in discrete_rot:
    freq_plot_rot(i)
    
LocID=df.groupby(['Location ID']).count()
country_count = LocID['Date'].value_counts()
sns.set(style="darkgrid")
sns.barplot(country_count.index[0:20],country_count.values[0:20],alpha=0.9)
plt.title('Number of Observations by '+'Location ID')
plt.ylabel('Number of Location IDs',fontsize=12)
plt.xlabel('Observations',fontsize=12)
plt.show()
```

Summary Statistics

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
print("\n Summary Statistics: \n")
print(df.drop(columns=['Location ID','Code','Promo']).describe())
```

Data cleaning - Check for missing values and adjust

```python
#checking NaN values
for col in df.columns:
    print("\n Column: ",col,", NaN values: ",df[col].isna().sum())
    
#checking inf values
for col in df.columns:
    if df[col].dtype!='O':
        print("\n Column: ",col,", Inf values: ",np.isinf(df[col]).values.sum())

#reassign Location Id
df['Location ID']=df.groupby(['Location ID']).ngroup()

# missing values adjustment:
spend=['Direct Mail Spend','Email Spend','TV Spend',
    'Affiliate Spend','Online Ad Spend','Facebook Ad Spend','Branded Search Spend'
            ,'Unbranded Search Spend'
            ,'Snapchat Spend'
            ,'Online Video Spend']
for i in spend:
    df[i].fillna(0)
categorical=['Code','Promo'] 
   
for i in categorical:
    df[i].fillna('Missing')
```

preprocessing for model estimation: logs and dummy variables

```python
for i in spend:
    df[i]=np.log(df[i])
    
df['Retail Transactions']=np.log(df['Retail Transactions'])
df['Web Transactions']=np.log(df['Web Transactions'])

# create dummy variables for year, month
year=pd.get_dummies(df['year'],drop_first=True,dummy_na=False).reset_index(drop=True)
year=year.add_prefix('year_')
month=pd.get_dummies(df['month'],drop_first=True,dummy_na=False).reset_index(drop=True)
month=month.add_prefix('month_')
df.drop(['year','month'],axis=1,inplace=True)
df=pd.concat([df,year,month],axis=1)
```

Estimation / Prediction Model for Web Transactions

```python
df_copy=df.copy()

df=df.set_index(['Date'])
df=df.set_index('Location ID',append=True)
# df=df.stack(level='Location ID')

dependent = 'Web Transactions'
explan=[
        # 'Number of Competitive Stores',
        'Number of Owned Stores',
        'Percentage Discount'
        ,'Affiliate Spend'
        ,'Email Spend'
        ,'Branded Search Spend'
        ,'Unbranded Search Spend'
        # ,'year'
        # ,'month'
        ]
extra=['Number of Competitive Stores'
       ,'year_2008'
       ,'year_2009'
       ,'year_2010'
       ,'year_2011'
       ,'month_2'
,'month_3'
,'month_4'
,'month_5'
,'month_6'
,'month_7'
,'month_8'
,'month_9'
,'month_10'
,'month_11'
,'month_12'       
       ]
both=list(explan.copy())
both.append(dependent)
both=both+extra
only_explan=list(explan.copy())
only_explan=explan+extra
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df=df[both].dropna()
dep=df[dependent]

# demean FE
# calculate the entity(state) mean beer tax
for i in explan:
    df['Mean_'+i] = df.groupby('Location ID')[i].transform(np.mean)
# calculate the entity(state) mean for fatal rate
df['Mean_'+dependent] = df.groupby('Location ID')[dependent].transform(np.mean)
# demean, subtract each row by the entity-mean
for i in explan:
    df[i] = df[i] - df['Mean_'+i]   
dep=dep-df['Mean_'+dependent]
dep=dep.astype(float)

exog = sm.tools.tools.add_constant(df[only_explan])    
for col in exog.columns:
    exog[col]=exog[col].astype(float)
model = sm.OLS(dep, exog)
results = model.fit(cov_type='HC3')
print(results.summary())
```

prediction of the model in-sample and fit with the data

```python
ypred = results.predict(exog)
for i in only_explan:
    if ('year' not in i) and ('month' not in i):
        scatter_fit_web(i)
```

Findings:
1. Web Transactions have time effects (year and month). Only year 2009 and month October are not statistically significant at the 5% significance level. Web Transactions increase in months of November and December, relative to January. The rest of the year, Web Transactions are lower relative to January (omitted). Year 2008 sees lower Web Transactions (holding everything else constant), and year 2009 doesn't have any statistical difference relative to year 2005 (omitted). Years 2010 and 2011 see larger Web Transactions. 
These year findings could reflect the economic crisis in 2008-2009 and how the Web Transactions are growing over time (years 2010-2011).   
2. Number of Competitive Stores is not statistically significant. This is not surprising as Web Transactions may belong to a segment of customers that is systematically different from the Retail sector.
3. Number of Owned Stores, Percentage Discount, Affiliate Spend, Email Spend, Branded Search Spend and Unbranded Search Spend all have a positive relation with Web Transactions and are statistically significant. Summary:
    * A 1% increase in Affiliate Spend is associated with a a 0.37% increase in Web Transactions, holding everything else constant.
    * A 1% increase in Email Spend is associated with a a 0.14% increase in Web Transactions, holding everything else constant.
    * A 1% increase in Unbranded Search Spend is associated with a a 0.09% increase in Web Transactions, holding everything else constant.
    * The above summary shows the importance of Digital advertising and the positive relation with Web Transactions.
    

Estimation / Prediction Model for Retail Transactions

```python
df=df_copy.copy()

df=df.set_index(['Date'])
df=df.set_index('Location ID',append=True)
# df=df.stack(level='Location ID')

dependent = 'Retail Transactions'
explan=[
        # 'Number of Competitive Stores',
        'Number of Owned Stores',
        'Percentage Discount'
        ,'Affiliate Spend'
        ,'Email Spend'
        ,'Branded Search Spend'
        ,'Unbranded Search Spend'
        # ,'year'
        # ,'month'
        ]
extra=['Number of Competitive Stores'
       ,'year_2008'
       ,'year_2009'
       ,'year_2010'
       ,'year_2011'
       ,'month_2'
,'month_3'
,'month_4'
,'month_5'
,'month_6'
,'month_7'
,'month_8'
,'month_9'
,'month_10'
,'month_11'
,'month_12'       
       ]
both=list(explan.copy())
both.append(dependent)
both=both+extra
only_explan=list(explan.copy())
only_explan=explan+extra
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df=df[both].dropna()
dep=df[dependent]

# demean FE
# calculate the entity(state) mean beer tax
for i in explan:
    df['Mean_'+i] = df.groupby('Location ID')[i].transform(np.mean)
# calculate the entity(state) mean for fatal rate
df['Mean_'+dependent] = df.groupby('Location ID')[dependent].transform(np.mean)
# demean, subtract each row by the entity-mean
for i in explan:
    df[i] = df[i] - df['Mean_'+i]   
dep=dep-df['Mean_'+dependent]
dep=dep.astype(float)

exog = sm.tools.tools.add_constant(df[only_explan])    
for col in exog.columns:
    exog[col]=exog[col].astype(float)
model = sm.OLS(dep, exog)
results = model.fit(cov_type='HC3')
print(results.summary())
```

prediction of the model in-sample and fit with the data

```python
ypred = results.predict(exog)

for i in only_explan:
    if ('year' not in i) and ('month' not in i):
        scatter_fit_retail(i)
```

Findings:
1. Web Transactions have time effects (year and month). All months show higher Retail Transations relative to January (omitted), statistically significant. However, there are seasonal effects. Months from May to August and November-December present higher Retail Transactions on average relative to the rest of the year. All years have higher Retail Transactions relative to 2005. The only exception is year 2008 with a negative coefficient, that is not statistically significant.   
2. Number of Competitive Stores is not statistically significant. However, Number of Owned Stores is statistically significant and it is positively related to Retail Transactions: an increase in 1 Owned Store is associated with a 1.3% increase in Retail Transactions.
3. Interestingly, Percentage Discount is negatively associated with Retail transactions. A 1% increase in Percentage Discount is associated with a 0.76% decrease in Retail Transactions. This finding is striking. It could be that Percentage Discount drives more traffic to Web Transactions (as found in the previous section), but it is less clear why that should drive lower Retail Transactions. It could be that firms know when they expect to sell less and therefore they offer a higher Percenage Discount in those periods, making the observed relationship biased/spurious (endogenous). It would be good to double check that the offered Percentage Discount can be redeemed both in Store and at the Web. 
4. Affiliate Spend, Email Spend, Branded Search Spend and Unbranded Search Spend all have a positive relation with Web Transactions and are statistically significant. Summary:
    * A 1% increase in Affiliate Spend is associated with a a 0.19% increase in Web Transactions, holding everything else constant.
    * A 1% increase in Email Spend is associated with a a 0.08% increase in Web Transactions, holding everything else constant.
    * A 1% increase in Unbranded Search Spend is associated with a a 0.04% increase in Web Transactions, holding everything else constant.
    * The above summary shows the importance of Digital advertising and the positive relation, not only with Web Transactions (as seen in the previous section), but Retail Transactions.
    

Final Recommendations:
<p>
<li>
    1. Digital advertising variables, such as Affiliate Spend, Email Spend, Branded Search Spend and Unbranded Search Spend are associated with higher both Web Transactions and Retail Transactions.
</li>
<li>
    2. Of those 4 channels, Affiliate Spend is the channel with the largest impact. The response to percent increases in any of these channels is largest in magnitude for Affiliate Spend.
</li>
<li>
    3. Firms would still ponder which channel is relatively more efficient for their specific marketing strategies.
</li>
<li>
    4. We would not recommend Percentage Discount as a means to increase Web Transactions as it is associated with lower Retail Transactions. At the very least, firms should ponder the trade-off that this action entails.
</li>
